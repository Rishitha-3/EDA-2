```python
#Discuss the scenarios where each scaling technique is preferred and why.
```

```python
#Min-Max Scaling

#Range: Typically [0, 1]

#When features have different units: Normalization scales all features to the same range, making them comparable.

#Algorithms sensitive to the scale of data: Algorithms like k-nearest neighbors (KNN), neural networks, and support vector machines (SVM) perform better when features are normalized because they rely on distance calculations.

#When using gradient-based optimizers: Normalization can speed up the convergence of gradient descent in algorithms like linear regression and neural networks.

#Why: It ensures that all features contribute equally to the model, preventing features with larger ranges from dominating the learning process.
```

```python
#Standard Scaling

#Range: Centered around 0 with a standard deviation of 1

#When features follow a Gaussian distribution: Standard scaling assumes a normal distribution and standardizes the features accordingly.

#When using algorithms that assume normality: Algorithms like logistic regression, linear discriminant analysis (LDA), and SVMs with RBF kernels assume that the input features are normally distributed.

#When features have outliers: Unlike min-max scaling, standard scaling is less sensitive to outliers because it centers the data around the mean.

#Why: It transforms the data to have zero mean and unit variance, making it suitable for algorithms that assume a normal distribution of features.
```

```python
#Min-Max Scaling: Use when features are on different scales and you need to bring them to a uniform range, especially for distance-based algorithms.

#Standard Scaling: Use when features are normally distributed and you need to standardize the data for algorithms assuming normality.
```

```python
#Discuss the pros and cons of One-Hot Encoding and Label Encoding.
```

```python
#One-Hot Encoding
#Pros:

#Non-Ordinal Data: One-Hot Encoding is suitable for categorical data where there is no intrinsic order among the categories (nominal data). It prevents the model from assuming any ordinal relationship.

#No Assumption of Ordinality: Since each category is represented by a binary vector, there is no assumption of any ordinal relationship, avoiding misleading interpretations.

#Compatibility with Algorithms: Many machine learning algorithms (e.g., linear regression, logistic regression) perform better with one-hot encoded features because it makes no assumptions about the order or magnitude of categories.

#Cons:

#High Dimensionality: One-Hot Encoding can lead to a significant increase in the dimensionality of the dataset, especially if the categorical feature has many unique values. This can result in the "curse of dimensionality" and lead to increased computational cost and memory usage.

#Sparse Data: The resulting binary vectors are sparse, containing many zeros, which can be inefficient to process and store.

#Redundancy: If one category can be inferred from the absence of others (e.g., if only three categories are represented by three binary features, the fourth category is implicit when all three features are zero), it can introduce redundancy.
```

```python
#Label Encoding
#Pros:

#Low Dimensionality: Label Encoding does not increase the dimensionality of the dataset, as it assigns an integer to each category.

#Efficiency: It is computationally efficient in terms of both memory and processing time since it simply replaces categories with integers.

#Simplicity: Easy to implement and interpret, as each category is assigned a unique integer value.

#Cons:

#Assumption of Ordinality: Label Encoding introduces an ordinal relationship between categories, which can mislead models into interpreting the encoded values as having a natural order or distance. This is inappropriate for non-ordinal data.

#Unintended Bias: The imposed ordinal relationship can lead to unintended biases in certain machine learning models, such as linear regression or distance-based algorithms, which might interpret the numeric values as ordered.

#Limited Applicability: Not suitable for algorithms that are sensitive to ordinal relationships if the categories do not have a natural order.
```

```python
import pandas as pd
import numpy as np
from scipy.stats import skew
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'adult_with_headers.csv'
data = pd.read_csv(file_path)

# Step 1: Create new features

# Feature 1: Capital gain minus capital loss
data['capital_gain_loss'] = data['capital_gain'] - data['capital_loss']

# Feature 2: Ratio of hours per week to years of education
data['hours_per_week_per_education_level'] = data['hours_per_week'] / data['education_num']

data.head()
```

```python
# Step 2: Check for skewness
numerical_features = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week', 'capital_gain_loss']
skewness = data[numerical_features].apply(lambda x: skew(x.dropna()))
print("Skewness before transformation:\n", skewness)
```

```python
# Step 3: Apply log transformation to highly skewed features (absolute skewness > 1)
skewed_features = skewness[abs(skewness) > 1].index
data[skewed_features] = data[skewed_features].apply(lambda x: np.log1p(x))

# Check skewness after transformation
skewness_after = data[skewed_features].apply(lambda x: skew(x.dropna()))
print("Skewness after transformation:\n", skewness_after)
```

```python
# Visualize the effect of log transformation on capital_gain
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(data['capital_gain'], kde=True)
plt.title('Capital Gain Distribution Before Log Transformation')

plt.subplot(1, 2, 2)
sns.histplot(np.log1p(data['capital_gain']), kde=True)
plt.title('Capital Gain Distribution After Log Transformation')

plt.show()
```

```python
# Visualize the effect of log transformation on capital_loss
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(data['capital_loss'], kde=True)
plt.title('Capital Gain Distribution Before Log Transformation')

plt.subplot(1, 2, 2)
sns.histplot(np.log1p(data['capital_loss']), kde=True)
plt.title('Capital Gain Distribution After Log Transformation')

plt.show()
```

```python
# Visualize the effect of log transformation
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(data['hours_per_week_per_education_level'], kde=True)
plt.title('Ratio of Hours/Week to Education Level Before Log Transformation')


plt.subplot(1, 2, 2)
sns.histplot(data['hours_per_week_per_education_level'], kde=True)
plt.title('Ratio of Hours/Week to Education Level After Log Transformation')


plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.ensemble import IsolationForest
import ppscore as pps
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'adult_with_headers.csv'
data = pd.read_csv(file_path)

# Step 1: Outlier Detection using Isolation Forest

# Select numerical features for outlier detection
numerical_features = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']

# Fit the Isolation Forest model
isolation_forest = IsolationForest(contamination=0.05, random_state=42)
data['outliers'] = isolation_forest.fit_predict(data[numerical_features])

# Remove outliers
data_cleaned = data[data['outliers'] == 1].drop(columns=['outliers'])
data_cleaned

```

```python
# Step 2: Calculate PPS matrix
pps_matrix = pps.matrix(data_cleaned)[['x', 'y', 'ppscore']].pivot(columns='y', index='x', values='ppscore')
pps_matrix.fillna(0, inplace=True)
pps_matrix
```

```python
# Calculate correlation matrix
correlation_matrix = data_cleaned.corr()
correlation_matrix
```

```python
# Visualize PPS matrix
plt.figure(figsize=(14, 10))
sns.heatmap(pps_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('PPS Matrix of Cleaned Dataset')
plt.show()
```

```python
# Visualize correlation matrix
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix of Cleaned Dataset')
plt.show()
```
